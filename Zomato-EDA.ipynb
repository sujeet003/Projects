```python
df_country=pd.read_excel("Country-Code.xlsx")
```

```python
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on ='Country Code',how="left")#takes left side)
#merges two DataFrame into single DataFrame
```

```python
final_df.head(2)
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#pie chart
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')#taking top 3 for more clearity
```

Observation: india has more share

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'RatingCount'})
#checks the size of particular columns,makes those to dataframe,rename and introduce a new columns and rename it
```

```python
ratings
```

Observations:
When rating is between 4.5 to 4.9 ---->Excellent

When ratings are between 5.0 to 4.4 ----->Very good

The ratring is between 3.4 to 3.9 ----->Good

The rating is between 2.5.0 to 3.4 ---->Average
             
The rating is between 1.8 to 2.4 ---->Poor

Zero is given from lot many users
    

```python
import matplotlib
sns.barplot(x='Aggregate rating',y="RatingCount",data=ratings)
matplotlib.rcParams['figure.figsize']=(12,6)# used to resize the any plots on matplotlib
```

```python
sns.barplot(x='Aggregate rating',y="RatingCount",data=ratings,hue="Rating color",palette=['blue',"red",'orange',"yellow","green","purple"])
```

Observation: Not rated nos are very high 

```python
#count plot(used to show categorical values)
sns.countplot(x="Rating color",data=ratings,palette=['blue',"red",'orange',"yellow","green","purple"])
```
